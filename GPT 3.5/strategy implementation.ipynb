```python
import pandas as pd
import numpy as np

green_cleaned = pd.read_csv('../green cleaned.csv', dtype={'ncusip': 'string'})
green_cleaned['ret_fwd_1'] = (green_cleaned.groupby('permno')['ret_excess'].shift(-1) )
```

```python
variables = ['datadate', 'permno', 'ticker', 'comnam', 'bm', 'mom12m', 'mve', 'ret_fwd_1']
```

```python
start_date = '2023-10-31'

analysis_df = green_cleaned[
    (green_cleaned['datadate'] == start_date)
][variables].copy()

conditions = [
    ( (analysis_df['bm'] > 1.5) & (analysis_df['mom12m'] > 1.0) ) | (analysis_df['mve'] > 1.5),
    
    # SELL: bm < -1.5 AND mom12m < -1.0
    ( (analysis_df['bm'] < -1.5) | (analysis_df['mom12m'] < -1.0) )
]

# 3. Define choices
choices = ['BUY', 'SELL']

# 4. Create the 'signal' column
analysis_df['signal'] = np.select(conditions, choices, default='HOLD')

buys = analysis_df[analysis_df['signal'] == 'BUY'][['datadate', 'permno', 'ticker', 'comnam', 'ret_fwd_1']]
sells = analysis_df[analysis_df['signal'] == 'SELL'][['datadate', 'permno', 'ticker', 'comnam', 'ret_fwd_1']]


print("\n--- BUY ---")
with pd.option_context('display.max_rows', None):
    print(buys)


print("\n--- SELL ---")
print(sells)

buys.to_csv('buys_2023.csv', index=False)
sells.to_csv('sells_2023.csv', index=False)
```

```python
start_date = '2023-12-31'

# BUY RULE:
# if (bm > 1.5 AND mom12m > 1.5) OR (bm > 1.5 AND mve < -1.5):
#     signal = BUY

# SELL RULE:
# if (bm < -1.5 OR mom12m < -1.5) AND mve > 1.5:
#     signal = SELL

analysis_df = green_cleaned[
    (green_cleaned['datadate'] == start_date)
][variables].copy()

conditions = [
    ( (analysis_df['bm'] > 1.5) & (analysis_df['mom12m'] > 1.5) ) | ((analysis_df['bm'] > 1.5) & (analysis_df['mve'] < -1.5) ),
    
    ( (analysis_df['bm'] < -1.5) | (analysis_df['mom12m'] < -1.5) ) & ( (analysis_df['mve'] > 1.5) )
]

# 3. Define choices
choices = ['BUY', 'SELL']

# 4. Create the 'signal' column
analysis_df['signal'] = np.select(conditions, choices, default='HOLD')

buys = analysis_df[analysis_df['signal'] == 'BUY'][['datadate', 'permno', 'ticker', 'comnam', 'ret_fwd_1']]
sells = analysis_df[analysis_df['signal'] == 'SELL'][['datadate', 'permno', 'ticker', 'comnam', 'ret_fwd_1']]


print("\n--- BUY ---")
with pd.option_context('display.max_rows', None):
    print(buys)


print("\n--- SELL ---")
print(sells)

buys.to_csv('buys_2024.csv', index=False)
sells.to_csv('sells_2024.csv', index=False)
```
